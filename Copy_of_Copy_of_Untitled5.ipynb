```
#Part I 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#load Data from pandas
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/haberman/haberman.data',header=None)
df.columns = ['Age_of_Patient', 'Patients_Year_of_Operation', 'Number_of_positive_axillary_nodes_detected', 'survival_status']
df.replace('?', np.NaN)
df.isna().sum()
df.head()





```

```
#Part II
pd.crosstab(df['Age_of_Patient'], df['survival_status'])
pd.crosstab(df['Patients_Year_of_Operation'], df['survival_status'])

```

```
#partII
plt.xlabel('Chances of surving Chancer')
plt.ylabel('Age of cancer patients')
plt.title('Does Age increases or decreases cancer?')
plt.bar(df['Age_of_Patient'], df['survival_status']);
```

```
#PartII
plt.xlabel('Year 1900s')
plt.ylabel('Years of living with cancer')
plt.title('If you were diagnosed with cancer during these years did your chance of survival decrease or increase?')
plt.bar(df['Patients_Year_of_Operation'], df['survival_status']);
```
