```
import pandas as pd #allows for cleaning/manipulating data
```

```
wine_data = pd.read_csv(wine_data_url, header=None) # assign var to pandas read_csv function to the adult census data url and eliminates the header (because it ends up
                                                    # using the first row of data as the category labels)
wine_data.dtypes
```

```
wine_data.shape
```

```
wine_data.columns = ['Cultivar 1-3','Alcohol','Malic Acid','Ash',
                     'Alcalinity of Ash','Magnesium','Total Phenols',
                     'Flavanoids','Nonflavanoid Phenols','Proanthocyanins',
                     'Color Intensity','Hue','OD280/OD315 of Diluted Wines','Proline']
```
