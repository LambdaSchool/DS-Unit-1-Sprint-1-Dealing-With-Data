```python
import seaborn as sns
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/inconvenient-sequel/ratings.csv')
```

```python
df.shape
```

```python
df.describe()
```

```python
pd.options.display.max_columns = None # To get all columns
df.head()
```

```python
df.dtypes
```

```python
df.timestamp = pd.to_datetime(df.timestamp)
df.set_index('timestamp', inplace = True)
```

```python
final = df[df.category == 'IMDb users']['respondents']
final.head()
```

```python
total_respondents = final.iloc[-1]
final = final * 100 / total_respondents
final.head()
```

```python
import matplotlib.ticker as ticker

plt.figure(figsize = (10, 7))
plt.title(s = "'An Inconvenient Sequel' was doomed before its release",
          x = 0,
          y = 1.11,
          ha = 'left')
plt.suptitle(t = 'Share of tickets sold and IMDb reviews for "An Inconvenient Sequel"\nposted through Aug. 27, by day',
             x = 0.09,
             y = 0.95,
             fontsize = 16,
             fontweight = 'ultralight',
             ha = 'left')
plt.style.use('fivethirtyeight')
ax = final.plot.line(color = '#47C0C7');

ax.set(xlabel = '',
       ylabel = '',
       yticks = range(0, 101, 20))
# ax.set_xlim(final.index[0])  # Gives error: 'Too many values to unpack'
# ax.set_xlim(auto = True)
ax.annotate('Share of IMDb\nreviews posted',
            (pd.to_datetime('2017-08-03'), 80),
            (pd.to_datetime('2017-08-03'), 80),
            fontweight = 'bold',
            color = '#47C0C7')

ax.annotate('60% of the reviews were already in\nby the wide release date',
            (pd.to_datetime('2017-08-05'), 61),
            (pd.to_datetime('2017-08-08'), 50),
            fontweight = 'ultralight',
            arrowprops=dict(facecolor='black', shrink=0.05))

# Change axis colors
ax.set_facecolor('w')

# dt.strftime("%A, %d. %B %Y %I:%M%p")

xlabels = ['', 'JULY 23, 2017', 'JULY 30', 'AUG. 6', 'AUG. 13', 'AUG. 20', 'AUG. 27']
@ticker.FuncFormatter
def major_formatter(x, pos):
  return xlabels[pos]
  
for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_horizontalalignment('center')
ax.xaxis.set_major_formatter(major_formatter)

# ax.get_xaxis().get_major_formatter().set_useOffset(False)

ax.tick_params(labelrotation = 0)
plt.show()
```
