```python
import numpy as np
import pandas as pd

#in order to prevent truncation set max rows and columns high
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
#we know that the dataframe has a row that is confused for a header
#we also know that the last two rows contain no information
#so we should skip those rows when reading the csv:
df = pd.read_csv('LoanStats_2018Q4.csv', skiprows=1, skipfooter=2, engine='python')
```

```python
#let's see how many instances and attributes we have:
print(df.shape)

#and get a feel for the data
df.head()
```

```python
#how many null values does this dataset contain?
df.isnull().sum().sort_values(ascending=False)
```

```python
#we know that there are some columns that have been scrubbed for privacy reasons
#so we'll get rid of those
df = df.drop(['id', 'url', 'desc', 'member_id'], axis='columns')
```

```python
#now let's strip the percentages from int_rate and convert them to floats

def remove_percent(string):
    return float(string.strip('%'))

df['int_rate'] = df['int_rate'].apply(remove_percent)

df.head()
```

```python
#now let's convert the 'term' column to an integer
#two strips, one to remove whitespace and another to remove months
def to_integer(string):
    return int(string.strip().strip('months'))

df['term'] = df['term'].apply(to_integer)

#by using .head we can confirm that we have successfully converted term
df['term'].head()
```

```python
#now let's add the new column called 'loan_status_is_great'
#criteria for being in this column is loan_status being current or fully paid

# the final condition in this func is just for in case you run the code twice
def loan_stat_int(string):
    if string == 'Current' or string == 'Fully Paid' or string == 1:
        return 1
    else:
        return 0
df['loan_status'] = df['loan_status'].apply(loan_stat_int)

df['loan_status'].head()
```

```python
#let's make sure that worked correctly
print((df['loan_status']==1).sum())

df_old = pd.read_csv('LoanStats_2018Q4.csv', skiprows=1, skipfooter=2, engine='python')
print((df_old['loan_status'] == 'Fully Paid').sum() + (df_old['loan_status'] == 'Current').sum())
```

```python
#now we'll make new columns for the last month and year a payment was made
df['last_pymnt_d'].head()
```

```python
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'])
```

```python
#let's check to make sure it's worked
df['last_pymnt_d'].head()
```

```python
#now we'll make the columns
df['last_pymnt_month'] = df['last_pymnt_d'].dt.month
df['last_pymnt_year'] = df['last_pymnt_d'].dt.year
```

```python
#and check again
df['last_pymnt_month'].head()
```
