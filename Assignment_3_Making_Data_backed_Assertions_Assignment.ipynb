```
for i in range(1500):
  exercise_time = random.uniform(20,400)
  age = random.random()>35+(exercise_time/1200)#if age is greater than 35 
  # out of 1200 people probably exercise time is low
  data_1.append(data(age,exercise_time,True))
```

```
for i in range(1300):
  exercise_time = random.uniform(10,350)              # if age is less than 35 exercise time is more
  age = random.random()<35+(exercise_time/1200)
  data_1.append(data(age,exercise_time,True))
  
```
