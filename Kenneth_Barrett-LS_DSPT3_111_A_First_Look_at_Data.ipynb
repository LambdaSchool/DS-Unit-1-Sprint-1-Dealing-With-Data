```
# I'm gonna try to style up a dataframe in pandas, because why not, right? :)

import pandas as pd

df = pd.read_csv("http://data.ctdata.org/dataset/9231ed6b-7682-43a8-b71a-c74e4034469b/resource/55660033-8d3d-4a0a-bcd5-87ed2ae1725c/download/marijuana-use-2016.csv")
```

```
# Cool, so the dataset has been imported! Let's make all the negative values red and positive green in column 'Value'

def red_or_green(val): # Defining the styling function here.
  color = 'red' if val < 0 else 'green'
  return 'color: %s' % color

df1 = df.style.applymap(red_or_green, subset='Value')


df1 # So it came out kind of weird. Wonder why that is?
```
