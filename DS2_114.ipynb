```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/wjarvis2/DS-Sprint-01-Dealing-With-Data/master/module4-databackedassertions/persons.csv')
print (df.head(50))
print (df.shape)
```

```
print ('Descriptive stats of age:')
print ('\n')
print (df['age'].describe())
print ('\n')       
print ('Descriptive stats of weight:')
print ('\n')
print (df['weight'].describe())
print ('\n')       
print ('Descriptive stats of exercise time:')
print ('\n')       
print (df['exercise_time'].describe())
```

```
# I think the most easily observable but relevant figures from these 3 sets of 
# descriptive stats are the min and max values of each column. We can tell right
# away that no person in the dataset was under 18 years old or over 80 years old.
# Weight has a min of 100 and a max of 246, which I think is actually pretty 
# light for a sample of 1200 people- especially considering the mean of only 153.
# Exercise time runs between 0 (minutes?) and 300. 
```

```
print ('A visual representation of age distribution:')
print ('\n')
df['age'].plot.density()
#df['age'].plot.hist()
plt.show();     
```

```
print ('A visual representation of weight distribution:')
print ('\n')
df['weight'].plot.density()
#df['weight'].plot.hist(bins=10)
plt.show();
```

```
print ('A visual representation of exercise time distribution:')
print ('\n')
df['exercise_time'].plot.density()
#df['exercise_time'].plot.hist(bins=10)
plt.show();
```

```
# After looking at these plots, it appears age looks fairly bi-modal, 
# Weight has a pretty considerable negative skew (likely because the min is 100)
# Exercise time looks decently normal
```

```
print ('Age column datatype:')
print (df['age'].dtypes)
print ('\n')
print ('Age column datatype:')
print (df['weight'].dtypes)
print ('\n')
print ('Age column datatype:')
print (df['exercise_time'].dtypes)
#The variable type for all 3 columns is int64
```
