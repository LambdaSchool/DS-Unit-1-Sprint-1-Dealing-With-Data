```python
df = pd.read_csv('LoanStats_2018Q4.csv', header = 1, skipfooter = 2, engine = 'python')
df.head()
```

```python
#Changing term into an integer
df['term']
```

```python
def term_to_int(x):
  return int(x[:-7])

df['term'] = df['term'].apply(term_to_int)
df.head()
```

```python
#Creating loan_status_is_great
df['loan_status_is_great'] = df['loan_status'].str.contains('Current') | df['loan_status'].str.contains('Fully Paid')
def bool_to_int(x):
  if x == True:
    return int(1)
  else:
    return int(0)
df['loan_status_is_great'] = df['loan_status_is_great'].apply(bool_to_int)
df.head()
```
