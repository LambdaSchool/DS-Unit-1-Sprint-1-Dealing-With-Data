```python
hardship_cols = [col for col in df.columns if col.startswith('hardship_')]
hardship_cols
```

```python
def remove_percent_to_float(x):
  return float(x[:-1])
```

```python
df['int_rate'] = df['int_rate'].apply(remove_percent_to_float)
```

```python
df.head()
```

```python
df['term']
```

```python

#ASSIGNMENT QUESTION CONVERT TERM COLUMN TO INTS#

df['term'] = df['term'].str.replace("months", "")
df['term']
```

```python
df['emp_title'].value_counts(dropna=False).head(20)
```

```python
df['emp_title'].nunique()
```

```python
def clean_title(title):
  if isinstance(title, str):
    return title.strip().lower()
  else:
    return 'unknown'
```

```python
df['emp_title'] = df['emp_title'].apply(clean_title)
```

```python
df['emp_title'].value_counts(dropna=False).head(20)
```

```python
df['emp_title_manager'] = df['emp_title'].str.contains('manager')
df.columns
```

```python

#ASSIGNMENT QUESTION CREATE 'loan_status_is_great' if it contains 'current' or 'fully paid' then display as binary 1 or 0 
df['loan_status_is_great'] = df['loan_status'].str.contains('Current', 'Fully Paid')
                                                      
```

```python
df['issue_year'] = df['issue_d'].dt.year
df['issue_month'] = df['issue_d'].dt.month
```

```python
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], infer_datetime_format=True)
```

```python
df['days_from_cr_line_to_loan'] = (df['issue_d'] - df['earliest_cr_line']).dt.days
```
