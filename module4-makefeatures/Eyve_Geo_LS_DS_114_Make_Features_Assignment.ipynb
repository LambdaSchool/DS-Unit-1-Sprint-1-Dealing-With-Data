```python
#clean up the titles 
def cleanTitle(item):
  if isinstance(item, str):
    return item.strip().title()
  else:
    return "Unknown"

df['emp_title'] = df['emp_title'].apply(cleanTitle)
#get a list of anything 'not in the top twenty'
topTwenty=df['emp_title'].value_counts(dropna=False).tail(-21)
bottomList=topTwenty.index.tolist()
len(bottomList)
```

```python
#this technically works, but I feel like there has to be a faster way. 
#this makes it go through and find each item in the list, it would be faster 
#to have it only check the top twenty list
df['emp_title']=df['emp_title'].replace(bottomList, "Other")
```

```python
#this is much more efficient and still works to do the same thing
topTwenty=df['emp_title'].value_counts(dropna=False).head(21)
topTwenty=topTwenty.index.tolist()
def otherTitles(item):
  if item in topTwenty:
    return item
  else:
    return 'Other'
#this line tells us about the results but doens't actually change anything

df['emp_title'].apply(otherTitles).value_counts()
```

```python
#down here it actually makes the change
df['emp_titles']=df['emp_title'].apply(otherTitles)
```
