```
import pandas as pd

#set the amount of rows/columns
pd.set_option('display.max_rows',50)
pd.set_option('display.max_columns',50)


#think about the header and the footheader
df_LoanStats = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter = 2)

```

# Dropped ID, memberId desc and URL
and loan_status_is_great function is here



```
import pandas as pd
import numpy as np
df_LoanStatsDropped=df_LoanStats.drop(columns=['id','member_id','desc','url'])
df_LoanStatsDropped.head(25)
def remove_percent_to_float(string):
  return float(string.strip('%'))

df_LoanStatsDropped['int_rate']=df_LoanStatsDropped['int_rate'].apply(remove_percent_to_float)
df_LoanStatsDropped




#######To do using pd.np.where(df[]. str.contains)
df_LoanStatsDropped["loan_status_is_great"] = pd.np.where(df_LoanStatsDropped["loan_status"].str.contains("Current"), "1",
                                                       pd.np.where(df_LoanStatsDropped["loan_status"].str.contains("Fully Paid."),"1","0"))
                                                       
#df_LoanStatsDropped["loan_status_is_great"]
#print(df_LoanStatsDropped["loan_status"])
df_LoanStatsDropped
##When I print out the dataframe the column header name is 0?
```

**if else statement for goal? and apply
**



```
#for i in df_LoanStatsDropped['loan_status']:
#  print(df_LoanStatsDropped['loan_status']=="Current")
def loan_status_is_great(string):
  if ((string=="Current") | (string=="Fully Paid.")):
    return int(1)
  else:
    return int(0)
#replacing all columns with apply function
df_LoanStatsDropped['loan_status_is_greats'] = df_LoanStatsDropped['loan_status'].apply(loan_status_is_great)
df_LoanStatsDropped['loan_status_is_greats']  #added at end of row. use .insert(x,y,value) to replace at a specific spot
```

Convert "term" to int

```
#replace all months with empty space
def remove_months_andConvert_float(string):
  return int(string.replace('months',''))

df_LoanStatsDropped['term'] = df_LoanStatsDropped['term'].apply(remove_months_andConvert_float)
df_LoanStatsDropped['term']
```

```
import datetime

df_LoanStatsDropped["issue_d"] = pd.to_datetime(df_LoanStatsDropped['issue_d'], infer_datetime_format=True)
df_LoanStatsDropped["issue_d"]
#let grab year
#dt mean format the date, only grab year
df_LoanStatsDropped['issue_d'].dt.year
#dt mean format the date, only grab month
df_LoanStatsDropped['issue_d'].dt.month

#add 36 months to the issue_d to get the last payment of the month
df_LoanStatsDropped["last_pymnt_date"]=df_LoanStatsDropped['issue_d'] + datetime.timedelta(36*365/12)
df_LoanStatsDropped["last_pymnt_d_month"]=df_LoanStatsDropped['last_pymnt_date'].dt.month
df_LoanStatsDropped["last_pymnt_d_month"]
#Yes the df which has been adjusted to obtained the year.
df_LoanStatsDropped["last_pymnt_d_year"]=df_LoanStatsDropped['last_pymnt_date'].dt.year
df_LoanStatsDropped["last_pymnt_d_year"]
```
