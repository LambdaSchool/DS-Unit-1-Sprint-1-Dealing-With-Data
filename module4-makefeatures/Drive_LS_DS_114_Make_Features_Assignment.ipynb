```python
# Finished Getting and Inspecting Data

# Next Load Data

import pandas as pd
df_Copy0 = pd.read_csv("LoanStats_2018Q4.csv.zip", header=1, skipfooter=2)
df_Copy0.head(40)
```

```python
# Convert term column from string to integer.

# Test Def

Term_Test = "36 months"
type(Term_Test)
```

```python
print(Term_Test.strip("months"))
```

```python
Term_Test = Term_Test.strip("months")
print(Term_Test)
type(Term_Test)
```

```python
Term_Test_int = int(Term_Test)
print(Term_Test_int)
type(Term_Test_int)
```

```python
def remove_months(string):
  return int(string.strip("months"))

remove_months(Term_Test)
```

```python
# Apply the function to the term column
df_Copy1 = df_Copy0
df_Copy1.term = df_Copy0.term.apply(remove_months)
df_Copy1.term
```

```python
# Convert terms to int
df_Copy1.term = df_Copy1.term.astype('int32')
df_Copy1.dtypes
```
