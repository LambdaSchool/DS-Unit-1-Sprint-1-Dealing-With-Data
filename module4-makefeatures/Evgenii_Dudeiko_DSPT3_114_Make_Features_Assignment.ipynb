##### There's one other column in the dataframe with percent signs. Remove them and convert to floats.

```python
df['int_rate'].isnull().sum()
```

```python
def remove_percent_to_float(string):
  return float(string.strip('%'))
```

```python
df['int_rate'] = df['int_rate'].apply(remove_percent_to_float)
```

##### Modify the emp_title column to replace titles with 'Other' if the title is not in the top 20.

```python
df['emp_title'].value_counts()[:20]
```

```python
df['emp_title'].isnull().sum()
```

```python
df['emp_title'].nunique()
```

```python
top_20 = df['emp_title'].value_counts().index[:20]
df['emp_title'] = np.where(df['emp_title'].isin(top_20), df['emp_title'], 'other')
valCount = df['emp_title'].value_counts()
```

```python
print (valCount)
```

```python
# make all the values lower
def clean_title(title):
  if isinstance(title, str):
    return title.strip().lower()
  else:
    return 'other'
```

```python
df['emp_title'] = df['emp_title'].apply(clean_title)
```

```python
df['emp_title'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,15))
series = df['emp_title'].value_counts()
labels = (np.array(series.index))
sizes = (np.array((series / series.sum())*100))
plt.pie(sizes, labels=labels, 
        autopct='%1.1f%%', startangle=200)
plt.title("Title distribution", fontsize=15)
plt.show()
```
