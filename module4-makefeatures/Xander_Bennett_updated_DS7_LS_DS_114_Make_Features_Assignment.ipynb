```python
# A list comprehension is essentially a 'for' loop being used on a list. That'd be good to use here

def remove_percent_to_float(string):
  return float(string.strip('%'))
```

```python
# Cleaning the title and handle missing values. What type are the NaN values?
import numpy as np
type(np.NaN)
```

```python
# Going to try an example of cleaning the emp_title data
examples = ['owner', 'Supervisor', ' Project Manager', np.NaN]

def clean_title(item):
  if isinstance(item, str):
    return item.strip().title()
  else:
    return 'Unknown'

[clean_title(item) for item in examples]
```

```python
# Now the NaN values are gone, and it looks cleaner. Now to use on the whole series
df['emp_title'] = df['emp_title'].apply(clean_title)

df.head()
```
