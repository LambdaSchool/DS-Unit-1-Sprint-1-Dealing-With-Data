```python
import pandas as pd

# Set Pandas Display Options
pd.set_option('display.max_rows', 400)
pd.set_option('display.max_columns', 400)


#header=1 is in de second line, if we write header= means header in the line 1
#skipfooter=2 means don't read last 2 rows
# python engine is currently more feature-complete
df = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2, engine='python') 

#print(rows,columns)
print(df.shape)
df.head()
```

```python
#check that the last 2 rows were deleted 
df.tail()
```

```python
#Sum allt the Null values and shows the result descending order
df.isnull().sum().sort_values(ascending=False)
```

```python
#there are 4 columns that are totally empty 
#in this case we are going to drop them because there is no data in the columns 
#Save the changes in a new data frame
df_new = df.drop(columns=['id', 'member_id', 'desc', 'url'], axis='columns') 
```

```python
def remove_month_to_int(string): #define function that is going to receive a string
  return int(string.strip('months'))
```

```python
#MAKE A TEST
```

```python
mlist=['6 months', '1234 months', '67 months']#define a list
```

```python
[remove_month_to_int(x) for x in mlist ] #apply the function for each item that is in mlist
```

```python
#we can create a new column an save the change there, but in this case we can to use the function apply() to save those changes
```

```python
df_new['term']= df_new['term'].apply(remove_month_to_int) #make the changes in the same column
#apply the function remove_month
```

```python
#2.  Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.
df_new['loan_status'].head(20)# show the column 
```

```python
df_new['loan_status'].unique()#List the unique values to see what are the categories 
```

```python
import numpy as np
```

```python
#I'm going to use the function np.where
# np.where(if condition , do, else) 
df_new['loan_status_is_great'] = np.where((df_new['loan_status'] == 'Current'), 1,np.where((df_new['loan_status'] == 'Fully Paid'), 1,0))
df_new[['loan_status_is_great','loan_status']].head(20)
```

```python
#3.  Make last_pymnt_d_month and last_pymnt_d_year columns.
```

```python
df_new['int_rate'].dtypes
```

```python
#Convert argument(integer, float, string, datetime, list, tuple, 1-d array, Series) to datetime.
df_new['last_pymnt_d'] = pd.to_datetime(df_new['last_pymnt_d'], infer_datetime_format=True)
```

```python
df_new['last_pymnt_d'].head(20)
```

```python
#create new column in which extract year from date and make it as int type
df_new['last_pymnt_d_year']=df_new['last_pymnt_d'].dt.year.astype('Int64')

#create new column in which extract month from date and make it as int type
df_new['last_pymnt_d_month']=df_new['last_pymnt_d'].dt.month.astype('Int64')

#shows three columns
df_new[['last_pymnt_d','last_pymnt_d_year','last_pymnt_d_month']].head(20)
```
