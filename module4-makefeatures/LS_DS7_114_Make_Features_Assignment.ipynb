```python
# Goal: Convert term column to one of integers
# Let's look at the term column
df['term'].head()
```

```python
# There are only two things to account for
df['term'].describe()
```

```python
# Convert to integer simply by taking the first characters after stripping whitespace
def termMonthToInt(x):
  return int(x.strip()[:2])
```

```python
df['term'] = df['term'].apply(termMonthToInt)
```

```python
df['term'].head()
```

```python
# Goal: Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.
# Let's look at loan status
df['loan_status'].head()
```

```python
# Create the list which we'll use as our new column
loanStatus = [1 if x in ['Current','Fully Paid'] else 0 for x in df['loan_status']]
```

```python
df['loan_status_is_great'] = loanStatus
```

```python
df['loan_status_is_great'].describe()
```

```python
# Check work
pd.crosstab(df['loan_status'], df['loan_status_is_great'])
```

```python
# Let's create a modified function to convert to floats while handling Nan values;
def convertPercentageString(x):
  if type(x) == str:
    return float(x.strip('%'))
  else:
    return np.NaN
```

```python
# Goal: Modify the emp_title column to replace titles with 'Other' if the title is not in the top 20. 
# First; let's get the top twenty jobs (not including the 21st 'Unknown' title)
top20 = list(df['emp_title'].value_counts().sort_values(ascending=False)[0:21].reset_index()['index'])
top20
```

```python
df['emp_title'] = [ x if x in top20 else 'Other' for x in df['emp_title']]
```
