```python
import pandas as pd
import numpy as np
```

```python
orig_df = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2, na_values=['n/a'], engine='python')
orig_df.head(20)
```

```python
# Run this block to Reset the Dataframe

df = orig_df.drop(columns=['id','member_id','desc','url'])
df.head()
```

```python
# Setting some Pandas display options for the dataframe
pd.set_option('display.max_rows', 450)
pd.set_option('display.max_columns', df.shape[1])
```

```python
df.dtypes
```

```python
df['term'].head(15)
```

```python
# ' month' Dropping function
def month_remover(string):
  return string.replace(' months', '')

df['term'] = df['term'].apply(month_remover)
```

```python
df['term'].head(15)
```

```python
df['term'] = pd.to_numeric(df['term'])
```

```python
df['term'].head(5)
```

```python
loan_status_list = ['Current', 'Fully Paid']
```

```python
df['loan_status_is_great'] = 0
df['loan_status_is_great'] = df['loan_status'].isin(loan_status_list) * 1
```

```python
df.head(4)
```

```python
# df.last_pymnt_d.head()
```

```python
# 

df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year
```

```python
df['last_pymnt_d_year']
```

```python
# Converting 'last_pymnt_d_<>' from a floating point, to a truncated int64

df['last_pymnt_d_month'] = pd.to_numeric(df['last_pymnt_d_month'], downcast='integer')
df['last_pymnt_d_year'] = pd.to_numeric(df['last_pymnt_d_year'], downcast='integer')
```
