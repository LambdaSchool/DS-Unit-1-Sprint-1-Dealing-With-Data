```python
df = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2)
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df = df.drop(columns=['id', 'member_id', 'desc', 'url'])
df.shape
```

```python
hardship_cols = [col for col in df.columns if col.startswith('hardship_')] #???
hardship_cols
```

```python
df = df.drop(columns=hardship_cols)
df.shape
```

```python
def remove_percent(x):
  return float(x[:-1])

df['int_rate'] = df['int_rate'].apply(remove_percent)

df.head()
```

```python
df['emp_title'].value_counts(dropna=False).head(20)
```

```python
def clean_title(title):
  if isinstance(title, str):
    return title.strip().lower()
  else:
    return 'unknown'
```

```python
df['emp_title'] = df['emp_title'].apply(clean_title)
df['emp_title'].value_counts(dropna = False).head(20)
```

```python
df['emp_title_manager'] = df['emp_title'].str.contains('manager')
df.head()
```

```python
mask = df['emp_title_manager'] == True
managers = df[mask]
plebians = df[~mask]

managers.shape

```

```python
plebians.shape
```

```python
assert plebians.shape[0] + managers.shape[0] == df.shape[0]
```

```python
df['issue_d'] = pd.to_datetime(df['issue_d'], infer_datetime_format=True)
df['issue_d']
```

```python
df['issue_year'] = df['issue_d'].dt.year
df['issue_month'] = df['issue_d'].dt.month
```

```python
df.head()
```

```python
df['issue_month']
```

```python
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], infer_datetime_format=True)
```

```python
df['days_to_loan'] = (df['issue_d'] - df['earliest_cr_line']).dt.days
df.head()
```

```python
df['issue_d'].dt.strftime('%Y, %B, %d')
```

```python
#Convert the `term` column from string to integer.

df['term']
```

```python
def remove_months(x):
  return int(x.strip(' months'))
```

```python
df['term'] = df['term'].apply(remove_months)
```
