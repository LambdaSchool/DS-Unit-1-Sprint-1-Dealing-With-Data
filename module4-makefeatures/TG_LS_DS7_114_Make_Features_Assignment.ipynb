###Clean the title and handle missing values

```python
import numpy as np

#NaN values are floats inside of Numpy
type(np.NaN)
```

```python
# Fill a list with a lower case, Title case, white space, and NaN value
examples = ['owner', 'Supervisor', ' Project Manager', np.NaN]

def clean_title(item): # Define function, takes a variable
  if isinstance(item, str): # If the variable is a string
    return item.strip().title() # Return the string title-formatted and strip white space.
  else: # If not a string var
    return 'Unknown'
  
[clean_title(item) for item in examples] #Clean the variables inside of examples
    
```

```python
# Apply clean_title function to all variables in 'emp_title' column
df['emp_title'] = df['emp_title'].apply(clean_title)

df.head()
```

###1- `term` column from string to integer

### 2 - Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.
