```python
## Convert int_"rate column to float from string

df['int_rate']
```

```python
## write fancy function
def remove_percent_to_float(x):
  return float(x[:-1])
```

```python
#call fancy function for all rows and assign back to int_rate
df['int_rate'] = df['int_rate'].apply(remove_percent_to_float)
```

```python
## Cleanup title

df['emp_title'].value_counts(dropna=False).head(30)
```

```python
# function to strip blanks, convert to lower case and if null setup sgtring "unknown"
def clean_title(title):
  if isinstance(title, str):
    return title.strip().lower()
  else:
    return 'unknown'
```

```python
df['emp_title'] = df['emp_title'].apply(clean_title)
```

```python
## Now lets do this : Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.


df['loan_status']
```

```python
## better write a func
def loan_status_validate(var):
  if (var == 'Current' or var == 'Fully Paid'):
    return (1)
  else:
    return(0)
```

```python
## test function

var = 'Late (31-120 days)'
print ( loan_status_validate (var))
```

```python
#now call function and create a new column

df['loan_status_is_great'] = df['loan_status'].apply(loan_status_validate)
```
