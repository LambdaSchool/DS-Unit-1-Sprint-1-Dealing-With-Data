```
#Stretch Goal
# ##Your Code Here
# Stretch Goals: Load the Dataset 1 into a colab notebook and answer the following questions. When you are finished, 
#add a link to your Colab notebook to the top of the code challenge as a comment.

# 1) Clean the data set and drop the Null/NaN values. Rename Product_Category_1-3 columns with an actual Product.

df_bf_cleaned = df_bf.dropna().rename(columns={
    'Product_Category_1': 'Trapezoidal_Picture_Frames',
    'Product_Category_2': 'One_Armed_Sweaters',
    'Product_Category_3': 'Extremely_Shiny_Mops'
  })
df_bf_cleaned.head(15)
```

```
# 2) How many unique user_ids does the data set contain?

print('Unique User_IDs: ' + str(df_bf_cleaned['User_ID'].nunique()))

# 3) How many unique age brackets are in the dataset. Which Age bracket has the most entries? Which has the least?

print('Unique age brackets: ' + str(df_bf_cleaned['Age'].nunique()))

# Conveniently, there aren't multiple modes/antimodes
# So we can do this the easy way
val_counts = df_bf_cleaned['Age'].value_counts()
print('Most frequent age bracket: ' + str(val_counts.index[0]))
print('Least frequent age bracket: ' + str(val_counts.index[-1]))


# 4) Transform the Gender categorical variable into a numerical variable. Then transform that numerical value into a Boolean.
df_bf_cleaned_num = df_bf_cleaned.copy()
df_bf_cleaned_num['Gender'].replace(to_replace={
    'F': 1,
    'M': 0
  }, inplace=True)
df_bf_cleaned_num.rename(columns={
    'Gender': 'Is_Female'
  }, inplace=True)
df_bf_cleaned_num.head()
```

```
# 4, cont'd

df_bf_cleaned_bool = df_bf_cleaned_num.copy()
df_bf_cleaned_bool['Is_Female'].replace(to_replace={
    0: False,
    1: True
  }, inplace=True)
df_bf_cleaned_bool.head()
```

```

# 5)What is the average Occupation score? What is the Standard Deviation? What is the maximum and minimum value?

print(df_bf_cleaned['Occupation'].describe())

# 7) Group Age by Gender and print out a cross tab with age as the y axis

pandas.crosstab(df_bf_cleaned['Age'], df_bf_cleaned['Gender'])
```

```
#Super Stretch Goal
##Your Code Here

from sklearn.metrics import mean_absolute_error, accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

model = LinearRegression()

train, test = train_test_split(df_bf_cleaned, random_state=0)
target = 'Purchase'
features = ['Trapezoidal_Picture_Frames', 'One_Armed_Sweaters', 'Extremely_Shiny_Mops', 'Occupation']

model.fit(train[features], train[target])

y_train_true = train[target]
y_train_predicted = model.predict(train[features])

# Accuracy score doesn't work for regression models
#print('Train accuracy:', accuracy_score(y_tr_true, y_tr_predicted))

# So instead, we'll use mean_absolute_error
train_error = mean_absolute_error(y_train_true, y_train_predicted)
print('Train error: ', train_error)



y_test_true = test[target]
y_test_predicted = model.predict(test[features])

test_error = mean_absolute_error(y_test_true, y_test_predicted)
print('Test error: ', test_error)
# Woohoo, didn't overfit too badly
# Though we'd probably get better results with a more complex model
# (And were we to fit off more features, but I'm out of time)
```
