```
# TODO - your code here
# Use what we did live in lecture as an example

import pandas as pd
import matplotlib.pyplot as plt

URL_person = 'https://raw.githubusercontent.com/DAVIDnHANG/DS-Unit-1-Sprint-1-Dealing-With-Data/master/module3-databackedassertions/persons.csv'
columns_name_person=('age','weight','exercise_time')
Person_CSV = pd.read_csv(URL_person, names = columns_name_person, header=1)
```

```
Weight_class = pd.cut(Person_CSV['weight'], 5)
ExerciseDuration = pd.cut(Person_CSV['exercise_time'],5)
AgeGroup = pd.cut(Person_CSV['age'],4)
crosstab = pd.crosstab(ExerciseDuration,[Weight_class,AgeGroup])
crosstab
crosstab_final = crosstab.iloc[1:5, 0:16 ]
crosstab_final
#crosstab_final.plot(kind='bar',figsize=(13,13))
```

```
#split the weight into two class. 
Weight_class = pd.cut(Person_CSV['weight'], 3)
ExerciseDuration = pd.cut(Person_CSV['exercise_time'],5)
AgeGroup = pd.cut(Person_CSV['age'],4)
crosstab = pd.crosstab(ExerciseDuration,[Weight_class,AgeGroup])
crosstab
crosstab_final = crosstab.iloc[1:5, 0:16 ]
crosstab_final
#crosstab_final.plot(kind='bar',figsize=(13,13))
```
