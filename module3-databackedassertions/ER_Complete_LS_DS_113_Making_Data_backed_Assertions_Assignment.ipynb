```
age_range = pd.cut(person_data['age'],3,precision=0)

# found exercise time to be a poor way of splitting the data, so I switched
# to age range instead.
# time_bins = pd.cut(person_data['exercise_time'],3)
pd.crosstab(person_data['weight'], age_range)
```

```
# checking to see if the data is clearer through normalizing columns, cause
# there's a lot

pd.crosstab(person_data['weight'], age_range, normalize='columns')

# nope...
```

```
ct = pd.crosstab(age_range, [person_data['exercise_time'],person_data['weight']])
ct
```

```
# Now I get irritated that it's all too large, I'm going to cut everything into
# bins!
weight_range = pd.cut(person_data['weight'],3,precision=0)
time_range = pd.cut(person_data['exercise_time'],5,precision=0)

ct2 = pd.crosstab(time_range, [age_range,weight_range])
ct2

# FINALLY! I can see and understand the data!
```

```
# My assertion from this data would be that, regardless of age, the thinnest 
# people will be able to workout longer than fatter people

# Now I want to see if I can make that come true on a graph

#ct2.plot(kind='bar') Nope

# ct2.plot() Still nope, but I think since age is not as important, it would
# make more sense to plot exercise time against weight here

ct3 = pd.crosstab(person_data['Unnamed: 0'],[age_range,time_range])
ct3.plot(kind='bar')
```
