Granted nobody can sustain a workout forever but in theory values of exercise_time are continuous. Exercise time can vary vastly and I feel would best be binned as shown in lecture.

3. I believe the relationships between age and exercise time as well as weight and exercise time are real relationships. Obviously this can not be proven until a linear regression or some sort of analysis is run on the data with a formal hypothesis;  but from a higher level point of view there appears to be some real relationship.
