```
weight_bins = pd.cut(df['weight'], 5, precision=0, include_lowest=True)   #precision=0 reduces the decimal places so I get whole number bins
age_bins = pd.cut(df['age'], 5, precision=0, include_lowest=True)
```

```
pd.crosstab(weight_bins, age_bins)
```

```
pd.crosstab(weight_bins, age_bins, normalize='columns')
```

```
weight_by_age = pd.crosstab(weight_bins, age_bins, normalize='columns')
```

```
weight_by_age_subset = weight_by_age.iloc[:, [0,4]]   #testing the theory that older people are going to be heavier on average
weight_by_age_subset.plot(kind='bar')
```

```
weight_by_age_subset.plot()
```

```
exercise_bins = pd.cut(df['exercise_time'], 5, precision=0)
```

```
pd.crosstab(weight_bins, exercise_bins, normalize='columns')
```

```
weight_by_exercise = pd.crosstab(weight_bins, exercise_bins, normalize='columns')
weight_by_exercise_subset = weight_by_exercise.iloc[:, [0, 2, 4]]
```

```
weight_by_exercise_subset.plot(kind='bar')   #hypothesis that both the low and high weight ends will exercise less than the middle
                                             #results do not bear this out, end up showing that the most exercise is done by the low weight bin
                                             #this might change if I increase the number of bins for weight groups - 100lbs to 130lbs includes both underweight and normal weight
```

```
weight_bins = pd.cut(df['weight'], 8, precision=0)
weight_by_exercise = pd.crosstab(weight_bins, exercise_bins, normalize='columns')
weight_by_exercise_subset = weight_by_exercise.iloc[:, [0, 2, 4]]
weight_by_exercise_subset.plot(kind='bar')
#this plot somewhat bears out my hypothesis - more people in the second bin exercise the longest than people in the first bin
```

```
weight_by_exercise_subset.plot()  #this graph shows the same thing more clearly, however there is still a stronger link between lower weight and more exercise than the difference
                                  #between the lowest and second lowest weight groups in exercise times
```

```
exercise_by_weight = pd.crosstab(exercise_bins, weight_bins, normalize='columns')  #does reversing the order of the variables change things?
exercise_by_weight_subset = exercise_by_weight.iloc[:, [0, 2, 4]]
exercise_by_weight_subset.plot(kind='bar')
```

1.  What are the variable types in the data?
Age in years is discrete numerical data. Weight in pounds and exericse time could be either continuous or discrete but this data set is dealing with whole number weights in pounds and exercise time in minutes, which means they are being treated as discrete in this instance. We can tell this because all of the data is the "int" type, which means it has to be whole numbers.

2.  What are the relationships between the variables?
It is more likely that someone in the higher age range will also be in the higher weight range, and vice versa. People in the lower weight ranges are in the higher exercise time ranges, but this isn't a perfect link - the people with lowest weights have slightly less time exercising than the people in the next weight group. People in the highest weight groups never reach the highest exercise times, but those in the lowest weight groups do reach the lowest exercise times. Exercise time and age start low together and climb steadily until about the middle of the data set for age, and then start to drop off again. Those in the highest age ranges never reach the highest exercise times. 

3.  Which relationships are "real", and which spurious?
