```
weight_bins = pd.cut(persons_data['weight'], 5)
age_bins = pd.cut(persons_data['age'], 10)


pd.crosstab(weight_bins, age_bins)
```

```
# ASK ABOUT SUMMING THESE COLUMNS AND ADDING A ROW 'TOTAL' AT THE END.
```

```
exercise_bins = pd.cut(persons_data['exercise_time'], 10)

pd.crosstab(exercise_bins, weight_bins)
```

```
pd.crosstab(exercise_bins, age_bins)
```

```
pd.crosstab(exercise_bins, (age_bins, weight_bins))
# Check on adding a third variable!!!
```
