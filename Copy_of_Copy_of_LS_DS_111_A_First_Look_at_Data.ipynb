```
# TODO - your code here
# Use what we did live in lecture as an example

import matplotlib.pyplot as plt
import numpy as np


fig, axs = plt.subplots(5, 5)
fig.set_figheight(15)
fig.set_figwidth(15)
colormap = ['plasma', 'viridis','inferno','magma','prism']
for col in range(5):
    for row in range(5):
        ax = axs[row, col]
        checkerboard = ax.pcolormesh(np.random.random((40, 40)) * (col + 1),
                            cmap=colormap[col])
        fig.colorbar(checkerboard, ax=ax)
plt.show()
```
