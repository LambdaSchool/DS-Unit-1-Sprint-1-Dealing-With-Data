```
import pandas as pd
Haberman_data = pd.read_csv(Haberman_data_url, header=None, engine='python')
Haberman_data.head(20)
```

```
# checking shape of data (number of observations)
Haberman_data.shape
```

```
# verifying all fields have values in the columns
Haberman_data.isnull().sum().sum()
```

```
# renaming column headers
Haberman_data.columns = ['Age of patient at time of operation (numerical)', 'Patients year of operation (year - 1900, numerical)', 'Number of positive axillary nodes detected (numerical)', 'Survival_status (class_attribute)_1 = patient_survived_5_years_or_longer, 2 = patient died within 5 year']
Haberman_data.head(10)
```

```
# changing ones and twos in last column to ones and zeros.
Haberman_data['Survival_status (class_attribute)_1 = patient_survived_5_years_or_longer, 2 = patient died within 5 year'] = Haberman_data['Survival_status (class_attribute)_1 = patient_survived_5_years_or_longer, 2 = patient died within 5 year'].replace(1, 0)
Haberman_data.head(10)
```

```
Haberman_data_2 = Haberman_data
Haberman_data_2['Survival_status (class_attribute)_1 = patient_survived_5_years_or_longer, 2 = patient died within 5 year'] = Haberman_data['Survival_status (class_attribute)_1 = patient_survived_5_years_or_longer, 2 = patient died within 5 year'].replace(2, 1)
Haberman_data_2.head(10)
```
