```python
Loans = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2)
Loans.head()
```

```python
def remove_string(x):
  return int(x[:3])
```

```python
remove_string('36 months')
```

```python
a = ['36 months', '60 months']
[remove_string(i) for i in a]
```

```python
Loans['term'] = Loans['term'].apply(remove_string)
```

```python
Loans.head()
```

```python
Loans.insert(144, 'loan_status_is_great', Loans['loan_status'], True)
```

```python
Loans.head()
```

```python
Loans['loan_status_is_great']
```

```python
Loans['loan_status_is_great'] = np.where(Loans['loan_status_is_great'].str.contains('Current'), 1,0) | np.where(Loans['loan_status_is_great'].str.contains('Fully Paid'), 1,0)
```

```python
Loans['loan_status_is_great']
```

```python
Loans.insert(145, 'last_pymnt_d_month', Loans['term'], True)
Loans.insert(146, 'last_pymnt_d_year', Loans['term'], True)
```
