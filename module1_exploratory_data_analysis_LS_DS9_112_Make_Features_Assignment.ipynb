##Convert int_rate

Define a functionto remove percent signs fromstrings and conver to floats

```python
def int_rate_to_float(mystring): # second int_rate is a parameter and can be any name
  return float(mystring.strip().strip('%')) # chaining left to right .strip().strip('%')
 # return float(mystring.strip().strip('%')) #/100
int_rate_to_float(int_rate) # second int_rate is the argument)
```
