####Dropping Columns

```
#DROP EMPTY COLUMNS
df = df.drop(columns = ['id', 'member_id', 'desc', 'url'])
df.head()
```

###Removing percents (%)

```
# def remove_percent_to_float(x):
#   return float(x[:-1])

remove_percent_to_float = lambda x:float(x[:-1])
```

```
df['int_rate'] = df['int_rate'].apply(remove_percent_to_float)
df.head()
```

####1. Convert the term column from string to integer

```
#TRYING A NEW FORMULA TO CHANGE TERM COLUMN FROM STRING TO INT
convert_str_to_int = lambda x:int(x[:-7])
```

```
df['term'] = df['term'].apply(convert_str_to_int)
df['term'].head(5)
```

####2. Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.

```
#DEFINE LOAN_STATUS_IS_GREAT
loan_status_is_great = df['loan_status'].str.contains('Current', 'Fully Paid')
loan_status_is_great.head(5)
```

```
#CREATE COLUMN HEADINGS
df['loan_status_is_great'] = loan_status_is_great
df.head()
```

```
#CONVERT FROM TRUE AND FALSE TO 0 AND 1
df['loan_status_is_great'] = df['loan_status_is_great']*1
df['loan_status_is_great'].head(5)
```

###3. Make last_pymnt_d_month and last_pymnt_d_year columns.

####Working With Dates

Converting the string format of issue_d to numbers


```
last_payment = pd.to_datetime(df['issue_d'], infer_datetime_format=True)
last_payment.head(5)
```

**Months**

```
# THE MONTHS WOULD BE THE SAME SINCE LOAN TERMS ARE BETWEEN 3 and 5 YEARS
df['last_pymnt_d_month'] = last_payment.dt.month
df.head()
```

**Years**

```
##FINDING THE VALUES IN THE TERM COLUMN TO ADD TO THE ISSUE MONTH AND YEAR
pd.value_counts(df['term'])
```

```
#NOW WORKING ON THE YEAR
last_pymnt_d_year = last_payment.dt.year
```

```
last_pymnt_d_year.head(5)
```

```
#ADDING 3 OR 5 TO THE YEARS COLUMN
last_pymnt_d_year = last_pymnt_d_year + df['term']/12
last_pymnt_d_year.astype(int).head(5)
```

```
#CREATE COLUMN HEADING
df['last_pymnt_d_year']=last_pymnt_d_year.astype(int)
df.head()
```

####2. Modify the emp_title column to replace titles with 'Other' if the title is not in the top 20.

Cleaning emp_title (from video)

```
df['emp_title'].value_counts(dropna=False).head(20)
```

```
df['emp_title'].nunique()
```

Clean the title and handle the missing values

```
def clean_title(title):
  if isinstance(title, str):
    return title.strip().lower()
  else:
    return 'unknown'
```

```
df['emp_title'] = df['emp_title'].apply(clean_title)
```

```
df['emp_title'].value_counts(dropna=False).head(20)
```

```
#TRYING TO NOT MANUALLY TYPE A LIST
top_20 = df['emp_title'].value_counts()[1:21].index.tolist()
top_20
```

```
def change_to_other(x):
  if x in top_20:
      return x
  else:
      return 'other'
```

```
others_column = df['emp_title'].apply(change_to_other)
```

```
others_column.value_counts()
```

```
df['emp_title'] = others_column
df['emp_title'].value_counts()
```
