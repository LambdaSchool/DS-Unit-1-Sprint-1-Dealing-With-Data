Let's **add it to a data frame**

```
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/bridges/bridges.data.version1', header = None)
df.columns = ['IDENTIF', 'RIVER', 'LOCATION', 'ERECTED', 'PURPOSE', 'LENGTH', 'LANES', 'CLEAR', 'TORD', 'MATERIAL', 'SPAN', 'REL_LENGTH', 'TYPE']
df.head()
```

**change '?'s to NaN**

```
nan_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/bridges/bridges.data.version1', header = None, na_values='?')
nan_df.columns = ['IDENTIF', 'RIVER', 'LOCATION', 'ERECTED', 'PURPOSE', 'LENGTH', 'LANES', 'CLEAR', 'TORD', 'MATERIAL', 'SPAN', 'REL_LENGTH', 'TYPE']
nan_df.head()
```

```
nan_df.isna().sum()
```

**Fill in NaNs**

Which ones could we use ffill for . . . LOCATION, CLEAR, T-OR-D, MATERIAL, SPAN, REL-LENGTH, TYPE

```
ffilled_df = nan_df.copy()
ffilled_df.head(20)
```

```
ffilled_df.CLEAR = ffilled_df.CLEAR.fillna(method='ffill')
ffilled_df.TORD = ffilled_df.TORD.fillna(method='ffill')
ffilled_df.MATERIAL = ffilled_df.MATERIAL.fillna(method='ffill')
ffilled_df.SPAN = ffilled_df.SPAN.fillna(method='ffill')
ffilled_df.REL_LENGTH = ffilled_df.REL_LENGTH.fillna(method='ffill')
ffilled_df.TYPE = ffilled_df.TYPE.fillna(method='ffill')
ffilled_df.LOCATION = ffilled_df.LOCATION.fillna(method='ffill')
ffilled_df.head()
```
