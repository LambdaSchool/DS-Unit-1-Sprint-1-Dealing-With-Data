## Loading LendingClub Data

```python
import pandas as pd

# Set pandas display options
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

df = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2, engine='python')
print(df.shape)
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

## Work with Strings

```python
# Let's begin the process of replacing the strings with numbers
df = df.drop(columns=['id', 'member_id', 'url', 'desc'], axis='columns')
```

```python
df.dtypes
```

```python
def split_term(string):
  return int(string.strip(' months'))
```

```python
sample_list = ['36 months', '48 months', '56 months']
[split_term(item) for item in sample_list]
```

```python
# .apply function now?
df['term'] = df['term'].apply(split_term)
```

```python
df.head()
```

```python
df.dtypes
# Took out output, but shows that it's an int now.
```

## Make Column: Loan Status is Great 

```python
df.loan_status_is_great = True
print(df.loan_status_is_great)
```

```python
df.dtypes
```

```python
# df['loan_status_is_great'] = df['loan_status'].str.contains("Current", "Fully Paid")
# df.head(5)
```

```python
df['loan_status_is_great'] = [1 if status in ['Current','Fully Paid'] else 0 for status in df['loan_status']]
df.head()
```

## Make `last_pymnt_d_month` and `last_pymnt_d_year` columns.

```python
# Let's convert the date and year on the last_pymnt_column by using to_datetime feature.
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)
```

```python
# Checking to see if it worked
df['last_pymnt_d'].head().values
```

```python
# df.dtypes confirms that it is now datetime64[ns]
df.dtypes
```

```python
# this shows us the month column on the last_pymnt_d column
last_pymnt_d_year = df['last_pymnt_d'].dt.month
```
