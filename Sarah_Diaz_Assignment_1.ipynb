```
import numpy as np
import seaborn as sns
sns.set(style="ticks")

rs = np.random.RandomState(10)
x = rs.gamma(2, size=1000)
y = -1 * x + rs.normal(size=1000)

sns.jointplot(x, y, kind="hex", color="#4c8cb3");
```
