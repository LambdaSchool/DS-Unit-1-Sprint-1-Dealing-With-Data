```
# download dataframe
import pandas as pd

# show more rows and columns
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
df = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2)
df.tail()
```

## 1. Convert the term column from string to integer.

```
# convert 'term' column from str to int (36 months -> '36')

def remove_months(string):
  return (string.strip('months'))
```

```
df['term'] = df['term'].apply(remove_months)
```

```
df['term'].tail()
```

```
df.tail()
```

## 2. Make a column named loan_status_is_great

```
df['loan_status_is_great'] = True

good_loan = ['Current', 'Fully Paid']

df['loan_status_is_great'] = df['loan_status'].isin(good_loan) * 1

df.head()
```

## 3. Make last_pymnt_d_month and last_pymnt_d_year columns

```
# view 'last_pymnt_d' column summary

df['last_pymnt_d'].describe()
```

```
# predicting mode
df['last_pymnt_d'].value_counts()
```

```
# checking NaN values sum

df['last_pymnt_d'].isnull().sum()
```

```
# replace NaN values with mode of the column

df['last_pymnt_d'].fillna(df.last_pymnt_d.mode(), inplace=True)
```

```
# converting 'last_pymnt_d' to datetime format

df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)
```

```
# create new column for 'last_pymnt_d_month'

df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month
```

```
# create new column for 'last_pymnt_d_year'

df['last_pymnt_d_year'] = True
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year
```
