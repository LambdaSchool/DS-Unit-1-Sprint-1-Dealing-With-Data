```
import pandas as pd

df = pd.read_csv(io.StringIO(uploaded['mars_express_test.csv'].decode('utf-8')))
```

```
df = df.rename(index=str, columns={
    'ut_ms': 'Unix Timestamp',
    'sa': 'Solar Aspect Angle',
    'sx': 'Solar Angle X',
    'sy': 'Solar Angle Y',
    'sz': 'Solar Angle Z'
    })

df.head(5)
```

```
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


for c, m, zlow, zhigh in [('r', '^', -100, -115), ('green', '^', -115, -130)]:
    xs = df['Solar Angle Y']
    ys = df['Solar Angle X']
    zs = df['Solar Angle Z']
    ax.scatter(xs, ys, zs, c=c, marker=m)

  
ax.set_xlabel('Solar Angle Y')
ax.set_ylabel('Solar Angle X')
ax.set_zlabel('Solar Angle Z')
plt.show();
```

```
plt.plot(df['Unix Timestamp'], df['Solar Aspect Angle'])
plt.plot;
```

```
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)


sns.scatterplot(x=df['Solar Angle X'], y=df['Solar Angle Z'],
                hue=df['Solar Angle Y'],
                palette="ch:r=-.2,d=.3_r",
                sizes=(1, 8), linewidth=0,
                data=df, ax=ax);
```
