## Modify the `emp_title` column

```python
df['emp_title'].value_counts()
```

```python
emp_title_value_count_df = pd.DataFrame()
```

```python
emp_titles = df['emp_title'].value_counts().index.tolist()
```

```python
emp_counts = df['emp_title'].value_counts().tolist()
```

```python
emp_title_value_count_df['emp_titles'] = emp_titles
emp_title_value_count_df['emp_counts'] = emp_counts

emp_title_value_count_df.head()
```

```python
emp_titles_to_change = []
```

```python
for index, row in emp_title_value_count_df.iterrows():
  if emp_title_value_count_df['emp_counts'][index] < 410:
    if emp_title_value_count_df['emp_titles'][index] not in emp_titles_to_change:
      emp_titles_to_change.append(emp_title_value_count_df['emp_titles'][index])
```

```python
emp_titles_to_change
```

```python
for index, row in df.iterrows():
  if df['emp_title'][index] in emp_titles_to_change:
    df['emp_title'][index] = "Other"
```
