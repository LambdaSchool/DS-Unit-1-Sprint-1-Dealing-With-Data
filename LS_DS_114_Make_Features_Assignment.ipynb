```python
import pandas as pd
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
df=pd.read_csv('LoanStats_2018Q4.csv', header=1, na_values=['n/a'],skipfooter=2)
df.head(20)
```

```python
list(df.columns)
```

```python
df.shape
```

```python
# List columns with total of missing values. Ascending order is default so have to reverse it

df.isnull().sum().sort_values(ascending=False)
```

```python
# Remove columns from the list

df=df.drop(columns=['id','member_id','desc','url'])
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Pull 'term' column
df['term']
```

```python
df.dtypes
```

```python
# Remove 'months' and remove space
term='36 months'
term.strip('months')[:-1]
```

```python
type(term.strip('months')[:-1])
```

```python
# Convert to float
float(term.strip('months')[:-1])
```

```python
type(float(term.strip('months')[:-1]))
```

```python
# Create function that works on any column
def remove_months_to_float(string):
  return float(string.strip('months')[:-1])

remove_months_to_float(term)
```

```python
df['term']=df['term'].apply(remove_months_to_float)
df.head()
```

```python
df.dtypes
```

```python
# Does 'loan_status' column has any missing values
df['loan_status'].isnull().sum()
```

```python
# Look up all loan status parameters
df['loan_status'].value_counts()
```

```python
#Create 'loan_status_is_great' column where 'Current' and 'Fully Paid' status marked as True
df['loan_status_is_great'] = True
df['loan_status_is_great']= df['loan_status'].str.contains('Current','Fully Paid')
df['loan_status_is_great']
```

```python
# Since True/False correspond to 1/0 convert to integer to reflect it

df['loan_status_is_great'] = df['loan_status_is_great'].astype(int)
df['loan_status_is_great']
```

```python

df['last_pymnt_d']
```

```python
# Convert to datetime format

df['last_pymnt_d']=pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)
```

```python
df.dtypes
```

```python
df['last_pymnt_d'].head().values
```

```python
df['last_pymnt_d'].dt.year
```

```python
df['last_pymnt_d_month']= df['last_pymnt_d'].dt.month
```

```python
df['last_pymnt_d_year']= df['last_pymnt_d'].dt.year
```
