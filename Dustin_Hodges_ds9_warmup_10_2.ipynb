## Pandas Practice

```
# Read a csv file from a specified source and print the first 5 rows
import pandas as pd
diamonds = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv')

print(diamonds.shape)
diamonds.head()
```

```
# Engineer a new feature named 'quality_color' using 'cut' and 'color', don't overthink it

diamonds['quality_cut'] = (diamonds['cut'] == "Ideal") & (diamonds['color'] == "J")
diamonds.head()
```

```
# Rename two of the columns
diamonds['length'] = diamonds['x']
diamonds['width'] = diamonds['y']

diamonds = diamonds.drop(['x', 'y'], axis=1)
diamonds.head()
```

```
# Use the 'carat' column to sort the dataframe, ascending
diamonds = diamonds.sort_values(by=['carat'])
diamonds.tail()
```

```
# Drop the 'carat' column from the dataframe

diamonds = diamonds.drop(['carat'], axis=1)
print(diamonds.shape)
diamonds.tail()
```

```
# Find diamonds where x (length) > 5, y (width) > 5, and z (depth) > 5
large = diamonds[(diamonds['length'] > 5) & (diamonds['width'] > 5) & (diamonds['z'] > 5)]
print(large.shape)
large.head()
```

```
# Filter the dataframe to display only diamonds that are 'premium' or 'ideal'
quality = diamonds[(diamonds['cut'] == "Ideal") | (diamonds['cut'] == "Premium")]
print(quality.shape)
quality.head()
```
