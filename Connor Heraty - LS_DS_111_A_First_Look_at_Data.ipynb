```
# Importing all the relevant python libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
# importing csv file from google drive

from google.colab import files
uploaded = files.upload()
```

```
df = pd.read_csv('alldata.csv')
```

```
# Checking for NaN values.
# There are some NaN values in the description feature, however we will not be using this and can ignore.

df.info()
```

```

df.head()
```

```
# My first visualization will plot the top 40 companies that are currently hiring data scientists.
# First we will need to convert our discrete variable 'position' into numerical format.


num_positions_bycompany = df.groupby(['company'])['position'].count()
num_positions_bycompany = num_positions_bycompany.reset_index(name= 'position')
num_positions_bycompany = num_positions_bycompany.sort_values(['position'],ascending=False)
num_positions_bycompany = num_positions_bycompany.head(40)
```

```
# I plotted the bargraph using matplotlib and seaborn

fig, ax = plt.subplots(figsize = (25,10))
ax = sns.barplot(x='company', y='position', data = num_positions_bycompany)
ax.set_xticklabels(num_positions_bycompany['company'], rotation=90)
ax.set_ylabel('Number of Open Positions', size=20)
ax.set_xlabel('Companies', size=20)
ax.set_title('Companies with the Most Demand for Data Scientists in the US (Aug 2018 - LinkedIn)', size=20);
```

```
df1 = pd.read_csv('alldata.csv')
```

```
# Since most of location observations include zip codes, I will replace all instances of the city/zip code to a streamlined format. This will make it easier to visualize later.

df1.location = df1.location.str.replace(r'((?i)^.*atlanta.*$)', 'Atlanta, GA')
df1.location = df1.location.str.replace(r'((?i)^.*austin.*$)', 'Austin, TX')
df1.location = df1.location.str.replace(r'((?i)^.*boulder.*$)', 'Boulder, CO')
df1.location = df1.location.str.replace(r'((?i)^.*boston.*$)', 'Boston, MA')
df1.location = df1.location.str.replace(r'((?i)^.*chicago.*$)', 'Chicago, IL')
df1.location = df1.location.str.replace(r'((?i)^.*washington.*$)', 'Washington, DC')
df1.location = df1.location.str.replace(r'((?i)^.*los Angeles.*$)', 'Los Angeles, CA')
df1.location = df1.location.str.replace(r'((?i)^.*cambridge.*$)', 'Cambridge, MA')
df1.location = df1.location.str.replace(r'((?i)^.*mountain view.*$)', 'Mountain View, CA')
df1.location = df1.location.str.replace(r'((?i)^.*new york.*$)', 'New York, NY')
df1.location = df1.location.str.replace(r'((?i)^.*redmond.*$)', 'Redmond, WA')
df1.location = df1.location.str.replace(r'((?i)^.*san diego.*$)', 'San Diego, CA')
df1.location = df1.location.str.replace(r'((?i)^.*seattle.*$)', 'Seattle, WA')
df1.location = df1.location.str.replace(r'((?i)^.*san francisco.*$)', 'San Francisco, CA')
df1.location = df1.location.str.replace(r'((?i)^.*sunnyvale.*$)', 'Sunnyvale, CA')


df1.head()
```

```
# This next visualization will show which cities currently have the most demand for Data Scientists
# As like above, we will need to convert our discrete variable 'position' into a numerical format.

num_open_byloc = df1.groupby(['location'])['position'].count()
num_open_byloc = num_open_byloc.reset_index(name= 'position')
num_open_byloc = num_open_byloc.sort_values(['position'],ascending=False)
openings_loc= num_open_byloc.head(40)
```

```
fig, ax = plt.subplots(figsize = (25,10))
ax = sns.barplot(x='location', y='position', data = openings_loc)
ax.set_xticklabels(openings_loc['location'], rotation=90)
ax.set_ylabel('Number of Open Positions', size=20)
ax.set_xlabel('Companies', size=20)
ax.set_title('US Cities with the Most Demand for Data Scientists (Aug 2018 - Linkedin)', size=20);
```

```
# This graph visualizes the most popular job titles demanded.

most_offd_roles=df.groupby(['position'])['company'].count()   
most_offd_roles=most_offd_roles.reset_index(name='company')
most_offd_roles=most_offd_roles.sort_values(['company'],ascending=False)
most_offd_roles=most_offd_roles.head(30)   


fig,ax=plt.subplots(figsize=(25,10))
ax=sns.barplot(x="position", y="company", data=most_offd_roles)    
ax.set_xticklabels(most_offd_roles['position'],rotation=90)
ax.set_xlabel('Most in Demand Data Science Jobs',fontsize=20)
ax.set_ylabel('Data Science Jobs',fontsize=20,)#
```
