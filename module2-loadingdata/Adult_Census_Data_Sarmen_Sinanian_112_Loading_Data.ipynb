```
df = pd.read_csv('https://raw.githubusercontent.com/ryanleeallred/datasets/master/adult.csv', na_values=' ?') #Set dataframe as variable df using pd (pandas) 
                                                                                                              # read_csv function to pull from URL
```

```
import numpy as np #Use numpy to manipulate objects in dataframe
df.replace(" ?", np.NaN, inplace=True) #Replace " ?" with NaN
df.head(20) #Print first 20 rows
```

```
df.isnull().sum() #Sum all NaN values within each column
```

```
df.shape #Show number of rows and columns
```

```
df2 = df.dropna() #Drop all rows with NaN
df2.head(20) #Row 14 shown as dropped
```

```
# df['female'] = (df.sex == 'female')
```

```
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x) #Strip whitespace from in front of and after each object/string
df.head()
```

```



# df.drop(['female'], axis=1)
df['female'] = (df.sex == 'Female').astype(int) #Create column named "female" and check column "sex" for values of "Female" then write a binary 1 in new column "female" for all "Female"
                                                # in existing "sex" column
                                                # and 0 for "Male" in new column "female" for all values of "Male" within column "female"
```

```
df[['sex','female']].head() #Check column "sex" and "female" for consistency
```

```
df.plot.scatter('female','capital-gain');
```

```
df.plot.scatter('female','capital-loss');
```

```
df['>50k'] = (df.salary == '>50K').astype(int)  #Create column named ">50k" and check column "salary" for values of ">50K" (capital K) then write a binary 1 in new
                                                # column ">50k" for all ">50K"
                                                # in existing "salary" column
                                                # and 0 for "<=50K" in new column ">50k" for all values of "<=50K" within column "salary"
```
