```
col_header = ['Indentifier', 'River', 'Location', 'Erected', 'Purpose', 'Length', 'Lanes', 'Clearing', 'T or D', 'Material', 'Span', 'Rel', 'Type']
bridge_data = pd.read_csv(bridge_data_url, header=None, names=col_header, na_values=['?'])
bridge_data.head()
```

```
bridge_data['River'] = bridge_data['River'].map({'M':'Monongahela', 'A':'Allegheny', 'O':'Ohio'})
bridge_data.head(10)
```

```
import numpy as np

bridge_data['Length'] = bridge_data['Length'].fillna(np.nanmedian(bridge_data['Length']))
bridge_data['Lanes'] = bridge_data['Lanes'].fillna(np.nanmedian(bridge_data['Lanes']))
bridge_data['Location'] = bridge_data['Location'].fillna(np.nanmedian(bridge_data['Location']))

print(bridge_data.isna().sum())
```
