# **Import Libraries**

```
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

# **Loading Data Set**

```

df2 = pd.read_csv('https://raw.githubusercontent.com/KryssyCo/DS-Unit-1-Sprint-5-Data-Storytelling-Blog-Post/master/national_afcars_trends_08thru17.updated2.csv')
print(df2.shape)
df2.head(10)
```

# **Cleaning and Exploring Data**

```
# Dropped two columns with unneccessary data.
df2 = df2.drop(['In Care on Sept 30th', 'Served'], axis=1)
```

```
# Check for NaN values
df2.isnull().sum()
```

```
# Check types of data in data frame
df2.dtypes
```

```
# # Change strings to ints and remove commas in numbers
df2['Adopted'] = df2['Adopted'].str.replace(',','').astype(int)
df2['Entered'] = df2['Entered'].str.replace(',','').astype(int)
df2['Exited'] = df2['Exited'].str.replace(',','').astype(int)
df2['Parental Rights Terminated'] = df2['Parental Rights Terminated'].str.replace(',','').astype(int)
df2['Waiting for Adoption'] = df2['Waiting for Adoption'].str.replace(',','').astype(int)
```

```
# Confirm data types were changed.
df2.dtypes
```

```
# Set index to Fiscal Years
df2.set_index('FY')
```

# **Plotting and Design Ideas**

```
# Designed and labeled plot

plt.style.use('fivethirtyeight');
fig = plt.figure(figsize=(8, 5));
df2.plot.bar()

plt.xticks(rotation='horizontal')
plt.xlabel('Years')
plt.ylabel('Amount of Children')
plt.title('Children languishing in the foster care system...still')
plt.show()
```

```
#Moved legend outside of plot
plt.style.use('fivethirtyeight');
# fig = plt.figure(figsize=(18, 15));
#df2.plot.bar(figsize=(8,7));
ax = df2.plot(colormap = 'cividis', kind='bar', title = 'Children in Foster Care',width = 1.025, figsize=(9,15));
plt.xticks(rotation='horizontal')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.xlabel('Years')
plt.ylabel('Amount of Children')
plt.title('Children languishing in the foster care system...still')
plt.show()
```

```
# Changed style and size of chart

ax = df2.plot(colormap = 'cividis', kind='barh', title = 'Children in Foster Care',width = 1.025, figsize=(9,15));

ax.set_yticks(df2.index)
ax.set_yticklabels(df2['FY'], rotation=45);
plt.xticks(rotation='horizontal')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.xlabel('Amount of Children');
plt.ylabel('Fiscal Years');
plt.title('Children in the foster care system...still')
plt.show()
```
