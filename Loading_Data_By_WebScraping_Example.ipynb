### The following loads an HTML table from a website into a pandas DataFrame

### This particular uses a Wikipedia Entry that lists people who have lived at airports (i.e. The Terminal)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


res = requests.get("https://en.wikipedia.org/wiki/List_of_people_who_have_lived_at_airports")
soup = BeautifulSoup(res.content)
table = soup.find_all('table')[0] 
df = pd.read_html(str(table))
```

```python
df_1 = df[0]
df_1
```

```python
df_1.drop(0, inplace=True)
df_1
```

```python
df_1.columns = ['Profile', 'Name', 'Nationality', 'Location', 'Period', 'Duration', 'Reason for stay', 'Reason for leaving']
```

```python
df_1.drop('Profile', axis=1, inplace=True)
```

```python
df_1
```

```python
df_1.to_csv('people_who_lived_at_airports.csv')
```
